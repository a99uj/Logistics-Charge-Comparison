```python
#importing the Pandas library to perform Data manipulation

import pandas as pd
```

```python
#A function to form 'weight slab (x)' column as per the min. & max. limit of the 'Total Weight (x)' column 

def weight_slab (weight):
    if weight <= 500:
        return 0.5
    elif weight <= 1000:
        return 1.0
    elif weight <= 1500:
        return 1.5
    elif weight <= 2000:
        return 2.0
    elif weight <= 2500:
        return 2.5
    elif weight <= 3000:
        return 3.0
    else:
        return 3.5
```

```python
#Applying the function to form the respective 'weight slab (x)' column

df_report_x['Weight Slab (x)'] = df_report_x['Total Weight (x)'].apply(lambda x: weight_slab(x))
```

```python
#Now, we need 'Expected Charge (x)' column -
#forming two columns 'Fixed  charge (x)' & 'Additional charge (x)' by the logic for that,
#they will deal with both 'Forward charges' & 'Forward and RTO charges' 
#accordingly set by the 'Zone (x)' & 'Type of Shipment' and the respective 'Weight Slab (x)'

#Function 'fixed charges' will deal with the first 0.5 KG weight of the item,
#Function 'additional charges' will deal with the each additional 0.5 KG weight in the same proportion.

def fixed_charges (zone, ship_type):
    if zone=='a' and ship_type=='Forward charges':
        return 29.5
    elif zone=='a' and ship_type=='Forward and RTO charges':
        return 29.5+13.6
    elif zone=='b' and ship_type=='Forward charges':
        return 33
    elif zone=='b' and ship_type=='Forward and RTO charges':
        return 33+20.5
    elif zone=='c' and ship_type=='Forward charges':
        return 40.1
    elif zone=='c' and ship_type=='Forward and RTO charges':
        return 40.1+31.9
    elif zone=='d' and ship_type=='Forward charges':
        return 45.4
    elif zone=='d' and ship_type=='Forward and RTO charges':
        return 45.4+41.3
    elif zone=='e' and ship_type=='Forward charges':
        return 56.6
    elif zone=='e' and ship_type=='Forward and RTO charges':
        return 56.6+50.7
#---
    
def additional_charges (weight,zone, ship_type):
    if weight>0.5:
        if zone=='a' and ship_type=='Forward charges':
            return 23.6
        elif zone=='a' and ship_type=='Forward and RTO charges':
            return 23.6+23.6
        elif zone=='b' and ship_type=='Forward charges':
            return 28.3
        elif zone=='b' and ship_type=='Forward and RTO charges':
            return 28.3+28.3
        elif zone=='c' and ship_type=='Forward charges':
            return 38.9
        elif zone=='c' and ship_type=='Forward and RTO charges':
            return 38.9+38.9
        elif zone=='d' and ship_type=='Forward charges':
            return 44.8
        elif zone=='d' and ship_type=='Forward and RTO charges':
            return 44.8+44.8
        elif zone=='e' and ship_type=='Forward charges':
            return 55.5
        elif zone=='e' and ship_type=='Forward and RTO charges':
            return 55.5+55.5
    else:
        return 1
```

```python
#Applying the function & forming the column 'Fixed charges (x)'

df_report_x['Fixed charges (x)'] = df_report_x.apply(lambda x: fixed_charges(x['Zone (x)'], x['Type of Shipment']), axis=1)
df_report_x['Fixed charges (x)'].round(2)
```

```python
#Applying the function & forming the column

df_report_x['Additional charges (x)'] = df_report_x.apply(lambda x: additional_charges(x['Weight Slab (x)'], x['Zone (x)'], x['Type of Shipment']), axis=1)
df_report_x['Additional charges (x)'].round(2)
```

```python
df_report_x
```

```python
#Using these two new columns to determine the 'Expected Charge (x)' column
#making a function for that as per the logic

def expected_charge (weight, fixed, additional):
    i=0
    j=0
    i = (weight/0.5) 
    j = i-1
    return (j*additional)+((i-j)*fixed)
```

```python
#Applying the function & forming the desired column

df_report_x['Expected Charge (x)'] = df_report_x.apply(lambda x: expected_charge(x['Weight Slab (x)'], x['Fixed charges (x)'], x['Additional charges (x)']), axis=1)
```

```python
df_report_x
```

```python
df_report_x['Billing Amount (Rs.)'].round(2)
```

```python
#Forming the final column 'Difference Between Expected Charges and Billed Charges (Rs.)'

def difference_in_charges (expected, actual):
    return float(expected - actual)

df_report_x['Difference Between Expected Charges and Billed Charges (Rs.)']= df_report_x.apply(lambda x: difference_in_charges(x['Expected Charge (x)'], x['Billing Amount (Rs.)']), axis=1)
```

```python
df_report_x['Difference Between Expected Charges and Billed Charges (Rs.)'].round(2)
```
